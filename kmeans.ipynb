```python
import pandas as pd
```

```python
df=pd.read_csv("wind_3.csv")
df.head(5)
```

```python
df.columns=df.columns.str.strip()
df.columns=df.columns.str.lower()
```

```python
df
```

```python
df['datetime'] = pd.to_datetime(df['date(yyyy-mm-dd)'] + ' ' + df['time (utc)'])
```

```python
df.drop(columns=['date(yyyy-mm-dd)', 'time (utc)'], inplace=True)
```

```python
df
```

```python
d1=df[["datetime","temp©","wind speed max","wind chill index"]]
d1
```

```python
d1=d1.set_index("datetime")
```

```python
d1
```

```python
d1['wind chill index']=pd.to_numeric(d1['wind chill index'],errors='coerce')
```

```python
d1.isnull().sum()
```

```python
d1["temp©"]=d1["temp©"].interpolate("linear")
```

```python
d1.isnull().sum()
```

```python
d1.dtypes
```

```python
from sklearn.cluster import KMeans
```

```python
features=["temp©","wind speed max","wind chill index"]
X=d1[features]
km=KMeans(n_clusters=3,random_state=42)
km.fit(X)
```

```python
X
```

```python
d1['Cluster'] = km.predict(X)
```

```python
d1
```

```python
import seaborn as sns 
sns.pairplot(d1, hue='Cluster')
```

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(X)
    sse.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(k_rng,sse)
plt.xlabel("k range")
plt.ylabel("sse")
plt.show()
```
